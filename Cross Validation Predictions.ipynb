```python
data = np.load('/home/philcox/Save/Nomoto_net/Hogg_medians3.1622776601683795.npz')
el = data['elements']
median = data['median']
upper = data['upper']
lower = data['lower']
data.close()
```

```python
parameter_names = [r'$\alpha_\mathrm{IMF}$',r'$\log_{10}(\mathrm{N_{Ia}})$',
           r'$\log_{10}(\mathrm{SFE})$',
           r'$\log_{10}(\mathrm{SFR_{peak}})$',r'$\mathrm{x}_{\mathrm{out}}$']
from Chempy.parameter import ModelParameters
a = ModelParameters()
prior = [] # Read prior  from parameter file
for i,param_name in enumerate(a.to_optimize):
    prior.append(a.priors.get(param_name)[0])
prior = np.array(prior)
sigma = [] # Read prior sigma from parameter file
for i,param_name in enumerate(a.to_optimize):
    sigma.append(a.priors.get(param_name)[1])
sigma = np.array(sigma)
```

```python
text_size = 22
plt.rc('font', family='serif',size = text_size)
plt.rc('xtick', labelsize=text_size)
plt.rc('ytick', labelsize=text_size)
plt.rc('axes', labelsize=text_size, lw=1.0)
plt.rc('lines', linewidth = 1)
plt.rcParams['ytick.major.pad']='8'
plt.rcParams['text.latex.preamble']=[r"\usepackage{libertine}"]
params = {'text.usetex' : True,
    'font.family' : 'libertine',
    'text.latex.unicode': True,
    }
plt.rcParams.update(params)
```

```python
fig,axes = plt.subplots(nrows = 5, ncols = 1,figsize=(40.,40.),dpi=300)
hspace = 0.35# blankspace height between subplots
plt.subplots_adjust(hspace=hspace)
plt.setp(axes, xticks=np.arange(len(el)), xticklabels=el)
for i in range(5):
    axes[i].errorbar(range(len(el)),median[:,i],yerr=[np.array(median[:,i])-np.array(lower[:,i]),np.array(upper[:,i])-np.array(median[:,i])],fmt='x',ms=10)
    axes[i].set_ylabel(str(parameter_names[i]))
    axes[i].set_xlabel(r"Cross-validation Element")
    axes[i].set_title(r"$\log_{10}{\beta} = 0.5$ inferences")
    axes[i].plot(np.arange(len(el)),np.ones(len(el))*prior[i],'k-')
    axes[i].plot(np.arange(len(el)),np.ones(len(el))*(prior[i]+sigma[i]),'k-.')
    axes[i].plot(np.arange(len(el)),np.ones(len(el))*(prior[i]-sigma[i]),'k-.')
plt.savefig('PaperPlots/CrossValidationPredictions.pdf')
```

Plots above show the posterior predictions (median / 1 sigma ranges) from the cross-validation MCMC, with the specified element removed. Dotted and dashed lines show the priors and 1-sigma prior ranges
