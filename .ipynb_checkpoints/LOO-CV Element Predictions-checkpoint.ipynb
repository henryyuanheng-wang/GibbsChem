```python
dat = np.load('/home/philcox/Save/OVERALL/Hogg_element_predictions_Nomoto2013_net.npz')
mean=dat['mean']
sigma=dat['sigma']
elements=dat['elements']
likelihood=dat['likelihood']
dat.close()

dat2 = np.load('/home/philcox/Save/OVERALL/Hogg_element_predictions_Chieffi04_net.npz')
mean2=dat2['mean']
sigma2=dat['sigma']
elements2=dat['elements']
likelihood2=dat['likelihood']
dat2.close()

# Load in proto-solar data
ps_dat = np.load('Chempy/input/stars/Proto-sun_all.npy')
ps_nam = ps_dat.dtype.names
ps_abun = []; ps_err = []
for el in elements:
    for i in range(len(ps_nam)):
        if ps_nam[i] == el:
            ps_abun.append(ps_dat[0][i])
            ps_err.append(ps_dat[1][i])
```

```python
plt.figure(figsize=(20,10))
text_size = 20
plt.rc('font', family='serif',size = text_size)
plt.rc('xtick', labelsize=text_size)
plt.rc('ytick', labelsize=text_size)
plt.rc('axes', labelsize=text_size, lw=1.0)
plt.rc('lines', linewidth = 1)
plt.rcParams['ytick.major.pad']='8'
plt.rcParams['text.latex.preamble']=[r"\usepackage{libertine}"]
params = {'text.usetex' : True,
    'font.family' : 'libertine',
    'text.latex.unicode': True,
    }
plt.rcParams.update(params)

# Iterate over trials
for i in range(9):
    plt.errorbar(np.arange(len(elements))+0.05*i*np.ones(len(elements)),mean[i],yerr=sigma[i],fmt='x',c='k',alpha=0.5)
    plt.errorbar(np.arange(len(elements))+0.05*i*np.ones(len(elements)),mean2[i],yerr=sigma2[i],fmt='o',c='b',alpha=0.5)
    
plt.errorbar(np.arange(len(elements))+0.05*9*np.ones(len(elements)),mean[9],yerr=sigma[9],fmt='x',c='k',alpha=0.5,label='Default')
plt.errorbar(np.arange(len(elements))+0.05*9*np.ones(len(elements)),mean2[9],yerr=sigma2[9],fmt='x',c='k',alpha=0.5,label='+Chieffi04')

plt.errorbar(np.arange(len(elements)),ps_abun,yerr=ps_err,fmt='^',c='r',label='Observations')
plt.ylabel('[X/Fe] abundance')
plt.xlabel('Cross Validation Element')
plt.legend()
ax=plt.gca()
la=plt.setp(ax,xticks=np.arange(len(elements)), xticklabels=elements)
```
