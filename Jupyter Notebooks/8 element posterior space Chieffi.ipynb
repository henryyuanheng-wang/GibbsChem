```python
from Chempy.plot_mcmc import restructure_chain
restructure_chain('mcmc/')
positions = np.load('mcmc/posteriorPDF.npy')
init_param = np.median(positions,axis=0)
```

```python
Nsigma = 3.
sigma = [] # Read prior sigma from parameter file
for i,param_name in enumerate(a.to_optimize):
    sigma.append(a.priors.get(param_name)[1])
sigma = np.array(sigma)
x=[]
y=[]
post=[]
#for i in range(2000):
  #  param = random.uniform(np.array(init_param)-Nsigma*np.array(sigma),np.array(init_param)+Nsigma*np.array(sigma),size=6)
 #   param[0] = init_param[0]
  #  param[3] = init_param[3]
  #  param[4] = init_param[4]
  #  param[5] = init_param[5]
  #  x.append(param[1])
  #  y.append(param[2])
  #  post.append(posterior(param))
  #  if i%500==0:
  #      print(i)
```

```python
from Chempy.cem_function import posterior_function_mcmc_quick
def posterior(theta):
    a = ModelParameters()
    post,_ = posterior_function_mcmc_quick(theta,a,preload)
    posterior = np.exp(post)
    return posterior
```

```python
ndata = 100
Nsigma=3
x = np.linspace(init_param[1]-Nsigma*sigma[1],init_param[1]+Nsigma*sigma[1], ndata)
y = np.linspace(init_param[2]-Nsigma*sigma[2],init_param[2]+Nsigma*sigma[2], ndata)
xi, yi = np.meshgrid(x, y)

theta = np.zeros((len(x),len(y)))
post = np.zeros((len(x),len(y)))

def post_output(x,y):
    for i in range(len(x)):
        if i %10==0:
            print(i)
        for j in range(len(y)):
            theta = [init_param[0],x[i],y[j],init_param[3],init_param[4],init_param[5]]
            tmp,_ = posterior_function_mcmc_quick(theta,preload.elements,preload)
            post[i,j] = np.exp(tmp)
    return post
    
Z = post_output(x,y)  
```

```python
print(a.to_optimize)
plt.clf()
fig = plt.figure(figsize=(8.69,6.69), dpi=300)
ax = fig.add_subplot(111)

### Tidy up plot
text_size=20
large_text = 30
label_size = 18
plt.rc('xtick', labelsize=text_size)
plt.rc('ytick', labelsize=text_size)
plt.rc('axes', labelsize=large_text, lw=1.)
plt.rc('lines', linewidth = 1.)
params = {'text.latex.preamble': r"\usepackage{libertine}",'text.usetex' : True,'font.size' : text_size,'font.family' : 'libertine','text.latex.unicode': True}
plt.rcParams.update(params)
###

plt.legend(loc='upper right')
plt.xlabel(r'$\alpha_\mathrm{IMF}$')
plt.ylabel(r'$\mathrm{log_{10}(N_{Ia})}$')
#plt.ylim([-0.02,1.08])
#plt.xlim([-0.0,1.0])

plt.contourf(x,y,np.log10(Z),200)
val = np.arange(-20,0,5)#-200,-60,30)
cbar = plt.colorbar()#ticks=val)
cbar.ax.set_title(r'$\log_{10}\mathrm{P}$')

# Compute covariance matrix for posterior
#positions = np.load('mcmc/posteriorPDF.npy')
#cov = np.zeros((2,2))
#for i in range(2):
#    for j in range(2):
#        cov[i,j] = np.cov((positions[:,i],positions[:,j]))[1,0]

# Plot prior contours
from matplotlib.patches import Ellipse
sig1 = Ellipse((a.p0[1],a.p0[2]),width=2*sigma[1],height=2*sigma[2],color='k',fill=False,ls='--',alpha=0.5)
sig2 = Ellipse((a.p0[1],a.p0[2]),width=4*sigma[1],height=4*sigma[2],color='k',fill=False,ls='--',alpha=0.5)
sig3 = Ellipse((a.p0[1],a.p0[2]),width=6*sigma[1],height=6*sigma[2],color='k',fill=False,ls='--',alpha=0.5)
ax.add_artist(sig1)
ax.add_artist(sig2)
ax.add_artist(sig3)

# Plot posterior peak and contours
plt.scatter(init_param[1], init_param[2],c='k',marker='x',s=30)
#for conf in [0.6827,0.9545,0.9973]:
#    plot_ellipse(cov=cov,ax=ax,x_cent=init_param[0],y_cent=init_param[1],mass_level=conf)


# Calculate 1,2,3 sigma contours
# Restructure inputs
in_dat = np.zeros((2,len(xi.ravel())))
for i in range(len(xi.ravel())):
    in_dat[0][i]=xi.ravel()[i]
    in_dat[1][i]=yi.ravel()[i]
out_dat = Z.ravel()

cov=np.cov(in_dat,aweights=out_dat)
rho = cov[0][1]/np.sqrt((cov[0][0]*cov[1][1])) # Find correlation between X and Y
sigma_conf = [3,2,1] # Confidences
f = np.exp(-0.5/(1-rho**2)*np.power(sigma_conf,2)) # Peak values

plt.contour(x, y, np.log10(Z)-np.max(np.log10(Z)),levels=np.log10(f),colors='k',linestyles='solid')

#fig.savefig('PaperPlots/PosteriorPlotbeta1.pdf',bbox_inches='tight')
```
