**Create a plot of the posterior function in parameter space taking points around the peak posterior value**

Do this for $\beta = 1$, which is the widest posterior we expect

```python
## First calculate the posterior position
from Chempy.wrapper import single_star_optimization
```

```python
# %load init.ipy
%pylab inline
from Chempy.parameter import ModelParameters
a = ModelParameters()
from Chempy.score_function import preload_params_mcmc
preload = preload_params_mcmc()
```

```python
single_star_optimization()
```

```python
from Chempy.plot_mcmc import restructure_chain
restructure_chain('mcmc/')
positions = np.load('mcmc/posteriorPDF.npy')
init_param = np.median(positions,axis=0)
```

```python
from Chempy.cem_function import posterior_function_mcmc_quick
def posterior(theta):
    a = ModelParameters()
    post,_ = posterior_function_mcmc_quick(theta,a,preload)
    posterior = np.exp(post)
    return posterior
```

```python
# Create random grid of points in 2 sigma width about centre changing 1st two parameters
Nsigma = 3.
sigma = [] # Read prior sigma from parameter file
for i,param_name in enumerate(a.to_optimize):
    sigma.append(a.priors.get(param_name)[1])
sigma = np.array(sigma)
x=[]
y=[]
post=[]
for i in range(2000):
    param = random.uniform(np.array(init_param)-Nsigma*np.array(sigma),np.array(init_param)+Nsigma*np.array(sigma),size=5)
    param[2] = init_param[2]
    param[3] = init_param[3]
    param[4] = init_param[4]
    x.append(param[0])
    y.append(param[1])
    post.append(posterior(param))
    if i%500==0:
        print(i)
```

```python
ndata = 100
x = np.linspace(init_param[0]-Nsigma*sigma[0],init_param[0]+Nsigma*sigma[0], ndata)
y = np.linspace(init_param[1]-Nsigma*sigma[1],init_param[1]+Nsigma*sigma[1], ndata)
xi, yi = np.meshgrid(x, y)

theta = np.zeros((len(x),len(y)))
post = np.zeros((len(x),len(y)))

def post_output(x,y):
    for i in range(len(x)):
        if i %10==0:
            print(i)
        for j in range(len(y)):
            theta = [x[i],y[j],init_param[2],init_param[3],init_param[4]]
            tmp,_ = posterior_function_mcmc_quick(theta,preload.elements,preload)
            post[i,j] = np.exp(tmp)
    return post
    
Z = post_output(x,y)  
```

```python
plt.clf()
fig = plt.figure(figsize=(8.69,6.69), dpi=300)
ax = fig.add_subplot(111)

### Tidy up plot
text_size=20
large_text = 30
label_size = 18
plt.rc('xtick', labelsize=text_size)
plt.rc('ytick', labelsize=text_size)
plt.rc('axes', labelsize=large_text, lw=1.)
plt.rc('lines', linewidth = 1.)
params = {'text.latex.preamble': r"\usepackage{libertine}",'text.usetex' : True,'font.size' : text_size,'font.family' : 'libertine','text.latex.unicode': True}
plt.rcParams.update(params)
###

plt.legend(loc='upper right')
plt.xlabel(r'$\alpha_\mathrm{IMF}$')
plt.ylabel(r'$\mathrm{log_{10}(N_{Ia})}$')
#plt.ylim([-0.02,1.08])
#plt.xlim([-0.0,1.0])

plt.contourf(x,y,np.log10(Z),200)
val = np.arange(-300,-40,20)
cbar = plt.colorbar(ticks=val)

from matplotlib.patches import Ellipse
sig1 = Ellipse((a.p0[0],a.p0[1]),width=2*sigma[0],height=2*sigma[1],color='k',fill=False,ls='--',alpha=0.5)
sig2 = Ellipse((a.p0[0],a.p0[1]),width=4*sigma[0],height=4*sigma[1],color='k',fill=False,ls='--',alpha=0.5)
sig3 = Ellipse((a.p0[0],a.p0[1]),width=6*sigma[0],height=6*sigma[1],color='k',fill=False,ls='--',alpha=0.5)
ax.add_artist(sig1)
ax.add_artist(sig2)
ax.add_artist(sig3)
plt.scatter(init_param[0], init_param[1],c='k',marker='x',s=30)

fig.savefig('PaperPlots/PosteriorPlotbeta1000.pdf',bbox_inches='tight')
```
