```python
%pylab inline
from Chempy.yields import SN2_feedback
chi = SN2_feedback()
getattr(chi,'chieffi04_net')()
tng = SN2_feedback()
getattr(tng,'TNG_net')()
```

```python
tng_tab = tng.table[tng.metallicities[-1]]
chi_tab = chi.table[0.02]

print('Property \t TNG \t C04')
print('------------------------------------------------')

for item in tng_tab.dtype.names:
    if item in chi_tab.dtype.names:
        print('%s \t\t %.2e \t\t %.2e' %(item,tng_tab[item][3],chi_tab[item][0]))
```
