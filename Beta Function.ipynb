```python
%pylab inline
from scipy.stats import beta
```

```python
x = linspace(0,1,500)
plt.figure()
list_of_beta_params = np.logspace(0,3,10) #[1,1.1,1.3,1.6,2,2.5,3.5,5,7,10,15,30,100,1000]
for a in list_of_beta_params:
    plt.plot(x,beta.pdf(x,1,a)/a,label='%.1f' %(a))
plt.legend(loc='upper right')
plt.rc('text',usetex=True)
plt.rc('font',family='serif')
plt.xlabel(r'$x$',fontsize=16)
plt.ylabel(r'$\beta(x)/k$',fontsize=16)
plt.ylim([1e-10,1.1])
plt.title(r'$\beta(x; 1, k)/k$ plot for various k',fontsize=16)
#plt.savefig('Presentation/beta.pdf',dpi=300)
```

```python
%pylab inline
x = np.linspace(0,10,100)
y = 0.19*x**2-0.5*x+2.5
y2 = 0.05*x**2-0.3*x+1
text_size=16
fig = plt.figure()
plt.clf()
plt.rc('font', family='serif',size = text_size)
plt.rc('xtick', labelsize=text_size/1.3)
plt.rc('ytick', labelsize=text_size/1.3)
plt.rc('axes', labelsize=text_size, lw=1.)
plt.rc('lines', linewidth = 1.)
plt.rcParams['ytick.major.pad']='8'
plt.rcParams['text.latex.preamble']=[r"\usepackage{libertine}"]
ax1=fig.add_subplot(111)
l1 = ax1.plot(x,y,label=r'$(\chi_1)$')
l2 = ax1.plot(x,y2,label=r'$\chi_2$')
ax2 = ax1.twinx()
l3 = ax2.plot(x,y/y2,label=r'$K$',c='r',ls='--')
lns = l1 + l2 + l3
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)
ax2.set_ylabel(r'$K_1/K_2$')
ax1.set_ylabel(r'$P(\mathcal{O}|\chi)$')
ax1.set_xlabel('k')
savefig('Kplot.pdf',dpi=300,bbox_inches='tight')
```
