```python
%pylab inline
from scipy.stats import beta
```

```python
plt.clf()
x = linspace(0,1,500)
fig = plt.figure(figsize=(16,8), dpi=300)
beta_log_param = [0.0,0.15,0.3,0.45,0.6,0.9,1.5,3]
list_of_beta_params = [10**item for item in beta_log_param] #np.logspace(0,3,20)
for i,a in enumerate(list_of_beta_params):
    plt.plot(x,beta.pdf(x,1,a)/a,label='%.2f' %(beta_log_param[i]))

### Tidy up plot
text_size=20
large_text = 30
plt.rc('xtick', labelsize=text_size)
plt.rc('ytick', labelsize=text_size)
plt.rc('axes', labelsize=large_text, lw=1.)
plt.rc('lines', linewidth = 1.)
params = {'text.latex.preamble': r"\usepackage{libertine}",'text.usetex' : True,'font.size' : text_size,'font.family' : 'libertine','text.latex.unicode': True}
plt.rcParams.update(params)
###

plt.legend(loc='upper right')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x;1,\beta)/\beta$')
plt.ylim([-0.05,1.05])
plt.savefig('PaperPlots/beta.pdf')
```

```python
%pylab inline
x = np.linspace(0,10,100)
y = 0.19*x**2-0.5*x+2.5
y2 = 0.05*x**2-0.3*x+1
text_size=16
fig = plt.figure()
plt.clf()
plt.rc('font', family='serif',size = text_size)
plt.rc('xtick', labelsize=text_size/1.3)
plt.rc('ytick', labelsize=text_size/1.3)
plt.rc('axes', labelsize=text_size, lw=1.)
plt.rc('lines', linewidth = 1.)
plt.rcParams['ytick.major.pad']='8'
plt.rcParams['text.latex.preamble']=[r"\usepackage{libertine}"]
ax1=fig.add_subplot(111)
l1 = ax1.plot(x,y,label=r'$(\chi_1)$')
l2 = ax1.plot(x,y2,label=r'$\chi_2$')
ax2 = ax1.twinx()
l3 = ax2.plot(x,y/y2,label=r'$K$',c='r',ls='--')
lns = l1 + l2 + l3
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)
ax2.set_ylabel(r'$K_1/K_2$')
ax1.set_ylabel(r'$P(\mathcal{O}|\chi)$')
ax1.set_xlabel('k')
savefig('Kplot.pdf',dpi=300,bbox_inches='tight')
```
